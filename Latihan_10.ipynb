```python
data_buah = {
    'apel' : 5000,
    'pisang' : 3000,
    'jeruk' : 7000,
    'anggur' : 15000
}
print(data_buah)
total = 0

for buah, harga in data_buah.items():
  print(f"harga {buah}: {harga}")
  total += harga
print(f"harga total: {total}")
```
