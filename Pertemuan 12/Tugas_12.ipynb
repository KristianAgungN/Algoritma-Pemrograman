```python
def luas_persegi_panjang(panjang, lebar):
  luas = panjang * lebar
  return luas
panjang = int(input("Masukkan panjang: "))
lebar = int(input("Masukkan lebar: "))
L = luas_persegi_panjang(panjang, lebar)
print(f'luas persegi panjang adalah = {L}\n')

def volume_persegi_panjang(panjang, lebar, tinggi):
  volume = panjang * lebar * tinggi
  return volume
panjang = int(input("Masukkan panjang: "))
lebar = int(input("Masukkan lebar: "))
tinggi = int(input("Masukkan tinggi: "))
V = volume_persegi_panjang(panjang, lebar, tinggi)
print(f'luas persegi panjang adalah = {V}\n')

def luas_lingkaran(jari_jari):
  luas = 3.14 * jari_jari ** 2
  return luas
jari_jari = int(input('Masukan jari-jari: '))
LL = luas_lingkaran(jari_jari)
print(f'Luas lingkaran adalah = {LL}\n')

def volume_bola(jari_jari):
  volume = 4/3 * 3,14 * jari_jari ** 3
  return volume
jari_jari = int(input('Masukan jari-jari: '))
VB = volume_bola(jari_jari)
print(f'Volume bola adalah = {VB}')
```
