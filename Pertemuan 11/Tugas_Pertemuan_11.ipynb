Buatlah program sistem manajemen perpustakaan sederhana (menambahkan dan
menampilkan buku baru ke dalam sistem) menggunakan konsep multi-keys dan nesting
dictionary dengan format
- KEYS
- JUDUL BUKU
- PENULIS
- TAHUN TERBIT
- TANGGAL TERAKHIR PINJAM
- STATUS (TERSEDIA / TIDAK)

```python
import datetime
import random
import string
from IPython.display import clear_output


buku_template = {
    'JB' : 'Judul Buku',
    'P' : 'Penulis',
    'TT' : 0000,
    'TTP' : datetime.datetime(1111,1,11),
    'S' : 'Status'
}

data_buku = {}

print(f'{'=====Perpustakaan=====' :^150}')

while True:
  buku = dict.fromkeys(buku_template.keys())

  buku['JB'] = input('Judul Buku = ')
  buku['P'] = input('Penulis = ')
  buku['TT'] = int(input('Tahun Terbit = '))
  buku['S'] = input('Tersedia / Kosong = ')

  Tahun_Pinjam = int(input('Tahun Pinjam (YYYY): '))
  Bulan_Pinjam = int(input('Bulan Pinjam (1-12): '))
  Tanggal_Pinjam = int(input('Tanggal Pinjam: '))


  buku['TTP'] = datetime.datetime(Tahun_Pinjam,Bulan_Pinjam,Tanggal_Pinjam)
  KEY = ''.join((random.choice(string.ascii_uppercase)for i in range(6)))

  clear_output(wait=True)

  print(f'{'=====Perpustakaan=====' :^150}')


  data_buku.update({KEY: buku})

  print(f'{'KEY':^8} {'Judul Buku':^35} {'Penulis':^25} {'Tahun Terbit':^20} {'Status':^20} {'Tanggal Terakhir Pinjam':^20} ')
  print('-'*140)

  for bp in data_buku:
    KEY = bp

    Judul_Buku = data_buku[KEY]['JB']
    Penulis = data_buku[KEY]['P']
    Tahun_Terbit = data_buku[KEY]['TT']
    Tanggal_Terakhir_Pinjam = data_buku[KEY]['TTP'].strftime('%d-%m-%y')
    Status = data_buku[KEY]['S']

    print(f'{KEY:^8} {Judul_Buku:^35} {Penulis:^25} {Tahun_Terbit:^20} {Status:^20} {Tanggal_Terakhir_Pinjam:^20} ')
    print('-'*140)
  print('\n')

  Add_data = input('Tambah Data? Y / N : ')
  if Add_data == 'N':
    break
```
